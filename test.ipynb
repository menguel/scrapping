```python
# Fonction pour parcourir récursivement l'arbre DOM

def traverse_dom(element, level=0):
    # Afficher l'element actuel
    if element.name:
        print(f"{' ' * level}<{element.name}>")
        
    # Si l'element a des enfants, les parcourir également
    
    if hasattr(element, 'children'):
        for child in element.children :
            traverse_dom(child, level + 1)
        
# Commencer le parcour depuis la racine de l'arbre DOM

traverse_dom(soup)
```
